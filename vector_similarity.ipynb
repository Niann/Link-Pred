```python
def sim(pair,tA,tB):
    vi,vj = pair
    tempA = np.zeros(l);
    tempB = np.zeros(l);
    tempA[tA[vi]] = 1/len(tA[vi]);
    for i in tB[vj]:
        tempB[tA[i]] += 1/len(tB[vj])/len(tA[i]);
    return cosine_similarity([tempA, tempB])[0][1]
```
