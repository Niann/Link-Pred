```python
def cossim(listA,listB):
    dot = len(list(set(listA).intersection(set(listB))))
    a = np.sqrt(len(listA))
    b = np.sqrt(len(listB))
    return dot/(a*b)
```

```python
predict = []
prepro = []

def pre(source,sink):
    sim = 0;
    j = 0;
    for i in Sink[sink]:
        j = j + 1;
        sim = sim + pow(cossim(Source[i],Source[source]),2.5)
    notsim = 0;
    for i in range(0,j):
        proxy = random.choice(list(Source))
        while (proxy in Sink[sink]):
            proxy = random.choice(list(Source))
        notsim = notsim + pow(cossim(Source[proxy],Source[source]),2.5)
    if not (sim < notsim):
        predict.append(1)
    else:
        predict.append(0)
    prepro.append(sim / j)
```
